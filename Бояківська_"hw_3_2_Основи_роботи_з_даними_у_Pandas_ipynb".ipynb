7.1. Використовуючи фільтрацію даних, знайдіть середнє значення `gross income` для міст `Yangon` та `Mandalay`. В якому місті воно більше?

7.2 Сформулюйте 3 гіпотези, чому в одному місті gross income вище, як можна дізнатись чи правдива ця гіпотеза з використанням даних (наявних і додаткових), та як висновки з цих гіпотез можна використати для прийняття рішень як підвищити прибуток в мережі супермаркетів. Відповідь напишіть текстом в окремій клітинці типу Markdown.

Це завдання - на продуктове мислення. Якщо не знаєте, які гіпотези сформулювати - напишіть ті, які здаються вам зовсім примітивними. Важливо навчитись висловлювати свої гіпотези щодо даних, і з часом та досвідом вони будуть ставати щоразу глибше і точніше. Весь аналіз даних будується на тому, що ми перевіряємо певні гіпотези та робимо висновки і рекоемндації на основі своїх досліджень.

```python
# Фільтрація даних для Yangon
yangon_income_mean = df[df['City'] == 'Yangon']['gross income'].mean()

# Фільтрація даних для Mandalay
mandalay_income_mean = df[df['City'] == 'Mandalay']['gross income'].mean()

# Виведення результатів
print(f"Середній дохід у Yangon: {yangon_income_mean:.2f}")
print(f"Середній дохід у Mandalay: {mandalay_income_mean:.2f}")

# Порівняння доходів
if yangon_income_mean > mandalay_income_mean:
    print("У місті Yangon середній дохід більший.")
else:
    print("У місті Mandalay середній дохід більший.")
```

Гіпотеза 1: Вищий середній чек
Гіпотеза: В одному з міст (наприклад, у Yangon) середній чек вищий, ніж у Mandalay. Це може бути через те, що в Yangon клієнти частіше купують дорожчі товари або купують більше товарів за один раз.

Як перевірити:

З використанням наявних даних можна обчислити середній чек у кожному місті, порівнявши показник Total у двох містах.
Додатково можна зібрати дані про асортимент товарів у кожному місті, щоб побачити, чи є різниця у наявності більш дорогих товарів.
Як використати:

Якщо виявиться, що в одному з міст середній чек значно вищий, варто переглянути асортимент товарів у інших містах і збільшити частку дорогих товарів.
Також можна провести рекламні кампанії для заохочення клієнтів купувати більше товарів за один раз.
Гіпотеза 2: Вища купівельна спроможність населення
Гіпотеза: У місті з вищим прибутком (наприклад, Yangon) серед клієнтів більше людей з вищою купівельною спроможністю. Це може пояснити вищий рівень продажів і, відповідно, прибутку.

Як перевірити:

З використанням наявних даних можна провести аналіз продажів за категоріями товарів і порівняти середню ціну товарів, які купуються в кожному місті.
Додаткові дані про середній дохід населення в кожному з міст допоможуть підтвердити або спростувати гіпотезу.
Як використати:

Якщо виявиться, що купівельна спроможність в одному місті вища, можна зосередитися на продажах товарів преміум-класу в цьому місті, а в інших містах проводити акції та знижки на більш доступні товари.
Гіпотеза 3: Різні рекламні стратегії
Гіпотеза: У місті з вищим прибутком використовуються ефективніші рекламні стратегії, які залучають більше клієнтів або стимулюють їх купувати більше.

9. Врахуовуючи висновок в попередньому завданні, дослідимо як продаються в цих містах різні категорії товарів. Виведіть відсотковий розпоіділ кількості інвойсів по категоріям товарів `Product line` в містах `Yangon` та `Mandalay`.

```python
df_yangon = df[df['City'] == 'Yangon']
df_mandalay = df[df['City'] == 'Mandalay']

# Відсотковий розподіл кількості інвойсів по категоріям товарів для Yangon
yangon_product_line_distribution = df_yangon['Product line'].value_counts(normalize=True) * 100

# Відсотковий розподіл кількості інвойсів по категоріям товарів для Mandalay
mandalay_product_line_distribution = df_mandalay['Product line'].value_counts(normalize=True) * 100

# Виведення результатів
print("Відсотковий розподіл кількості інвойсів по категоріям товарів у місті Yangon:")
print(yangon_product_line_distribution)
print("\nВідсотковий розподіл кількості інвойсів по категоріям товарів у місті Mandalay:")
print(mandalay_product_line_distribution)
```
