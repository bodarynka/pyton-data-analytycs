- **atemp**: як відчувається температура в градусах Цельсія

Використайте бібліотеку Seaborn для створення **парних графіків** (pairplot) розсіювання, які показуватимуть взаємозв'язки між кількістю зареєстрованих, неформальних користувачів та загальною кількістю велосипедів, взятих на прокат. Кожен графік повинен розрізняти робочі та святкові дні за кольором.

Для побудови цієї візуалізації спочатку відфільтруйте потрібні колонки в даних і потім зробіть візуалізацію.

Проаналізуйте візуалізацію і скажіть, чи можна виявити особливу поведінку користувачів у святкові дні з цих графіків?

```python
filtered_df = df[['registered', 'casual', 'count', 'workingday']]


sns.pairplot(filtered_df, hue='workingday', palette='Set1')

plt.suptitle("Взаємозв'язки між кількістю користувачів і загальною кількістю прокатів", y=1.02)

plt.show()
```

 Тепер давайте візуалізуємо кількість зареєстрованих та казуальних користувачів поденно. Для цього виконайте код нижче і після цього створіть нову змінну `df_viz` яка містить лише колонки `'date', 'registered', 'casual'`, згрупуйте дані за датою і відобразіть на **одному** графіку поденну динаміку двох показників з допомогою `pandas.plot`.

```python
df['date'] = df.datetime.dt.date
```

```python
df_viz = df[['date', 'registered', 'casual']]

# Згрупуємо дані за датою та підсумуємо значення для кожної дати
df_viz = df_viz.groupby('date').sum()

# Відобразимо поденну динаміку зареєстрованих та незареєстрованих користувачів
plt.figure(figsize=(10, 8))
df_viz.plot(y=['registered', 'casual'], figsize=(10, 5))
plt.title('Поденна динаміка зареєстрованих та незареєстрованих користувачів')
plt.xlabel('Дата')
plt.ylabel('Кількість користувачів')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(['registred', 'casual'])
plt.show()
```
