0. Імпортуйте бібліотеку pandas та зчитайте дані у змінну `df` типу `pandas.DataFrame`.

```python
import pandas as pd
df = pd.read_csv('supermarket_sales.csv')
```

1. Дослідимо, який філіал супермаркету ('Branch') є найприбутковішим. Для цього знайдіть сумарний прибуток за кожним філіалом і виявіть, який філіал має найвищий.

```python
file_path = '/content/supermarket_sales.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
branch_profit = df.groupby('Branch')['gross income'].sum()


print(branch_profit)

# Знаходимо філіал із найвищим прибутком
most_profitable_branch = branch_profit.idxmax()
highest_profit = branch_profit.max()

print(f"Найприбутковіший філіал: {most_profitable_branch} з прибутком {highest_profit:.2f}")


```

2.2. В якому місті знайходиться філіал з найвищим прибутком? Може в тому місці нам розмітисти ще один магазин.  


```python
city_of_profitable_branch = df[df['Branch'] == most_profitable_branch]['City'].unique()[0]

print(f"Філіал з найвищим прибутком знаходиться в місті {city_of_profitable_branch}.")
```

3.1. Створіть зводну таблицю, яка покаже, скільки покупок (інвойсів) було зроблено в кожній з філій (`Branch`) за різними категоріями товарів. Запишіть таблицю в змінну `invoices_by_category` і виведіть змінну на екран.
Ця таблиця допоможе проаналізувати, в якій філії купують найбільше товарів кожної з категорій.

```python
invoices_by_category = pd.pivot_table(
    df,
    index='Branch',           # Рядки: Філії
    columns='Product line',   # Стовпці: Категорії товарів
    values='Invoice ID',      # Поле для підрахунку: Ідентифікатор інвойсу
    aggfunc='count',          # Функція агрегації: підрахунок інвойсів
    fill_value=0              # Заповнюємо відсутні значення нулями
)

invoices_by_category
```

```python
max_electronics_branch = invoices_by_category['Electronic accessories'].idxmax()
max_electronics_branch_count = invoices_by_category['Electronic accessories'].max()

print(f"Філіал з найбільшою кількістю інвойсів у категорії 'Electronic accessories': {max_electronics_branch}")
print(f"Кількість інвойсів: {max_electronics_branch_count}")
```

**Гіпотеза 1:** Вплив часу покупки на вибір товарів

Питання: У які години дня найчастіше купують конкретні категорії товарів, наприклад, їжу та напої або електронні аксесуари?

Як перевірити: Використати зведені таблиці та функцію groupby для аналізу розподілу покупок у різних часових інтервалах (годинах) для кожної категорії товарів.


```python
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour

# Створимо зведену таблицю для кількості покупок у кожній годині за категоріями товарів
category_by_hour = df.groupby(['Hour', 'Product line']).size().unstack().fillna(0)

# Виведемо кількість покупок для категорій 'Food and beverages' та 'Electronic accessories'
category_by_hour[['Food and beverages', 'Electronic accessories']]
```

З створеної таблиці бачимо, що найчастіше здійснюються покупки з категорії Food and beverages, а саме о 10-й, 15-й та з 19-20-й годині. Та в категорії Electronic accessories з 10-11 години та о 18-й.

**Гіпотеза2 :** Вплив типу клієнтів на середній чек

Питання: Чи відрізняється середній чек у клієнтів-членів супермаркету від звичайних клієнтів і в яких категоріях товарів ця різниця найбільша?

```python
# Обчислимо середній чек для різних типів клієнтів за категоріями товарів
average_check_by_customer_type = df.groupby(['Product line', 'Customer type'])['Total'].mean().unstack()

# Виведемо результат
average_check_by_customer_type
```


*Рекомендації*:

 Персоналізувати пропозиції для членів на основі категорій, де вони витрачають більше, щоб *підвищити* їхню активність.

Розробити нові програми лояльності або пропозиції, спрямовані на збільшення витрат у категоріях, де різниця в середньому чеку *найбільша*.

# Гіпотеза3: Вплив методу оплати на обсяги покупок

Питання: Чи різняться обсяги покупок (кількість інвойсів) залежно від методу оплати, і якщо так, то для яких товарних категорій це найпомітніше?

```python
# Обчислимо кількість інвойсів за методом оплати і категоріями товарів
invoices_by_payment_method = df.groupby(['Product line', 'Payment']).size().unstack().fillna(0)

# Виведемо таблицю, яка показує розподіл кількості інвойсів за методами оплати
invoices_by_payment_method
```
